Task 1.0: Setup and Image Loading

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image and convert to RGB for Matplotlib display
# We do this once to be used in all sub-tasks
img = cv2.imread('imagesLab3\\cameraman.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get original dimensions for later use
height, width = img.shape[:2]

if img is None:
    print("Error: Image not found! Check the path.")
else:
    print("Image loaded successfully.")
```

Task 1.1: Increase the size of the image (Scaling)

```python
# Apply Scaling (1.5x larger)
new_size = (int(width * 1.5), int(height * 1.5))
resized_img = cv2.resize(img_rgb, new_size, interpolation=cv2.INTER_CUBIC)

# Display the Original Image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img_rgb)

# Display the Resized Image with zoom 1.5x
plt.subplot(1, 2, 2)
plt.title('Resized Image (1.5x)')
plt.imshow(resized_img)
plt.show()
```

Task 1.2: Rotate the image by 120 degrees

```python
# Define rotation matrix
center = (width // 2, height // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 120, 1.0)

# Apply rotation
rotated_img = cv2.warpAffine(img_rgb, rotation_matrix, (width, height))

# Display the Original Image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img_rgb)

#Display the Rotated Image with 120 Degrees
plt.subplot(1, 2, 2)
plt.title('Rotated 120 Degrees')
plt.imshow(rotated_img)
plt.show()
```

Task 1.3: Perform Shear operations

```python
# Define Shear matrix
shear_factor = 0.2
shear_matrix = np.float32([[1, shear_factor, 0], [0, 1, 0]])

# Apply Shear
new_width = int(width + abs(shear_factor * height))
sheared_img = cv2.warpAffine(img_rgb, shear_matrix, (new_width, height))

# Display the Original Image
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img_rgb)

# Display the Sheared Image
plt.subplot(1, 2, 2)
plt.title('Sheared Image')
plt.imshow(sheared_img)
plt.show()
```
