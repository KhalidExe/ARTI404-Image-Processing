Task 2.0: Setup and Image Loading

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image once in Grayscale for all intensity tasks
# We use '0' because intensity transformations are applied to gray levels
img = cv2.imread('imagesLab3\\cameraman.png', 0)

if img is None:
    print("Error: Image not found! Check the path.")
else:
    print("Image loaded successfully.")
```

Task 2.1: Negative Transformation

```python
# Apply Negative Transformation: s = 255 - r
img_negative = 255 - img

# Display the Original Image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original (Grayscale)')
plt.imshow(img, cmap='gray')

# Display the Negative Image "Opposite"
plt.subplot(1, 2, 2)
plt.title('Negative Image')
plt.imshow(img_negative, cmap='gray')
plt.show()
```

Task 2.2: Log Transformation

```python
# Calculate the constant 'c' to normalize the result to [0, 255]
c = 255 / np.log(1 + np.max(img))
# Apply Log Transformation formula: s = c * log(1 + r)
log_img = c * (np.log(img + 1.0))

# Convert the result to uint8 format for display
log_img = np.array(log_img, dtype=np.uint8)

# Display the Original Image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')

# Display the Log Transformation Image
plt.subplot(1, 2, 2)
plt.title('Log Transformation')
plt.imshow(log_img, cmap='gray')
plt.show()
```

Task 2.3: Power Law (Gamma) Transformation

```python
# Gamma > 1 increases contrast in bright areas
gamma = 2.0 
# Apply Power Law
gamma_img = np.array(255 * (img / 255) ** gamma, dtype='uint8')

# Display the Original Image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(img, cmap='gray')

# Display the Power Law Image
plt.subplot(1, 2, 2)
plt.title(f'Power Law (Gamma={gamma})')
plt.imshow(gamma_img, cmap='gray')
plt.show()
```
